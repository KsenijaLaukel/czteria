Tutaj pojawia się wiele wyzwań. Nie chcę twierdzić, że sztuczna inteligencja jest aż tak słaba, ale uważam, że nie jest jeszcze na poziomie, który pozwoliłby w pełni zastąpić człowieka w takich zadaniach.  

*  odpowiadanie na pytania klientów w internetowej infolinii `+`, myślę, że sztuczna inteligencja może sobie z tym dobrze poradzić, ponieważ odpowiedzi na pisemne pytania opierają się na określonych, szablonowych odpowiedziach, które nie wymagają głębokiej analizy.
