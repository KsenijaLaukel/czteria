```python
#1
Fakty
:- discontiguous mezczyzna/1.
:- discontiguous kobieta/1.
% Relacje rodzic-dziecko
ojciec(jan, anna).
ojciec(jan, piotr).
matka(maria, anna).
matka(maria, piotr).

% Relacje małżeńskie
malzenstwo(jan, maria).

% Płeć
mezczyzna(jan).
kobieta(maria).
mezczyzna(piotr).
kobieta(anna).

Reguły
% Przodek(X, Y) - X jest przodkiem Y
przodek(X, Y) :-
    ojciec(X, Y).
przodek(X, Y) :-
    matka(X, Y).
przodek(X, Y) :-
    ojciec(X, Z),
    przodek(Z, Y).
przodek(X, Y) :-
    matka(X, Z),
    przodek(Z, Y).

% Brat(X, Y) - X jest bratem Y
brat(X, Y) :-
    mezczyzna(X),
    ojciec(Z, X),
    ojciec(Z, Y),
    X \= Y.

% Siostra(X, Y) - X jest siostrą Y
siostra(X, Y) :-
    kobieta(X),
    matka(Z, X),
    matka(Z, Y),
    X \= Y.

% Kuzyn(X, Y) - X jest kuzynem Y
kuzyn(X, Y) :-
    ojciec(A, X),
    ojciec(B, Y),
    malzenstwo(A, C),
    malzenstwo(B, C),
    X \= Y.

Zapytanie
przodek(jan, anna).

Result
true

Zapytanie
brat(piotr, anna).

Result
true

Zapytanie
kuzyn(X, piotr).

Result
anna
```

```python
#2
:- dynamic lotnisko/1.
:- dynamic polaczenie/3.
Fakty:
lotnisko(warszawa).
lotnisko(krakow).
lotnisko(gdansk).

polaczenie(warszawa, krakow, 300).
polaczenie(krakow, gdansk, 200).
polaczenie(warszawa, gdansk, 500).

Reguły
polaczenie(X, Y) :-
    polaczenie(X, Y, _);
    polaczenie(Y, X, _).

bezposrednie_polaczenie(X, Y) :-
    polaczenie(X, Y).

koszt_podrozy(X, Y, C) :-
    polaczenie(X, Y, C);
    polaczenie(Y, X, C).

najkrotsza_droga(X, Y) :-
    najkrotsza_droga(X, Y, Trasa, MinimalnyKoszt),
    format('Trasa: ~w, Koszt: ~w', [Trasa, MinimalnyKoszt]).

najkrotsza_droga(X, Y, Trasa, MinimalnyKoszt) :-
    znajdz_wszystkie_trasy(X, Y, WszystkieTrasy),
    znajdz_koszty(WszystkieTrasy, Koszty),
    minimalny_koszt(WszystkieTrasy, Koszty, Trasa, MinimalnyKoszt).

znajdz_wszystkie_trasy(X, Y, Trasy) :-
    setof([Sciezka, Koszt], Sciezka^Koszt^dfs(X, Y, [X], Sciezka, Koszt), Trasy).

dfs(X, Y, Trasa, [Y|Trasa], Koszt) :-
    polaczenie(X, Y, Koszt).
dfs(X, Y, Wizyta, Sciezka, Koszt) :-
    polaczenie(X, Z, C),
    Z \= Y,
    \+ member(Z, Wizyta),
    dfs(Z, Y, [Z|Wizyta], Sciezka, Koszt2),
    Koszt is C + Koszt2.

znajdz_koszty([], []).
znajdz_koszty([[_, Koszt]|ResztaTras], [Koszt|ResztaKosztow]) :-
    znajdz_koszty(ResztaTras, ResztaKosztow).

minimalny_koszt([Trasa|_], [Koszt|_], Trasa, Koszt).
minimalny_koszt([_|ResztaTras], [_|ResztaKosztow], Trasa, Koszt) :-
    minimalny_koszt(ResztaTras, ResztaKosztow, Trasa, Koszt).

Zapytanie
polaczenie(warszawa, krakow).

Result
true

Zapytanie
koszt_podrozy(warszawa, gdansk, C).

Result
C = 500

Zapytanie
najkrotsza_droga(warszawa, gdansk).

Result
Trasa: [[gdansk,krakow,warszawa],500], Koszt: 500
true
```

```python
#3
Fakty:
:- dynamic produkt/3.
produkt(tv, 10, sektor_a).
produkt(pralka, 5, sektor_b).
produkt(lodowka, 3, sektor_c).

Reguły
dostepny_produkt(X) :-
    produkt(X, Ilosc, _),
    Ilosc > 0.

przenies_produkt(X, NowySektor) :-
    produkt(X, Ilosc, StarySektor),
    retract(produkt(X, Ilosc, StarySektor)),
    assertz(produkt(X, Ilosc, NowySektor)),
    format('produkt(~w, ~w, ~w)', [X, Ilosc, NowySektor]).

uzupelnij_stan(X, N) :-
    produkt(X, Ilosc, Sektor),
    NewIlosc is Ilosc + N,
    retract(produkt(X, Ilosc, Sektor)),
    assertz(produkt(X, NewIlosc, Sektor)),
    format('produkt(~w, ~w, ~w)', [X, NewIlosc, Sektor]).

Zapytanie
dostepny_produkt(tv).

Result
true

Zapytanie
przenies_produkt(tv, sektor_b).

Result
produkt(tv, 10, sektor_b)
true

Zapytanie
uzupelnij_stan(pralka, 3).

Result
produkt(pralka, 8, sektor_b)
true
```

```python
#4
Fakty
miasto(warszawa).
miasto(krakow).
miasto(poznan).
miasto(wroclaw).
miasto(lodz).
miasto(szczecin)

droga(warszawa, krakow, 4).
droga(krakow, poznan, 3).
droga(warszawa, poznan, 7).
droga(warszawa, wroclaw, 5).
droga(wroclaw, poznan, 2).
droga(krakow, wroclaw, 2).
droga(lodz, krakow, 5).
droga(lodz, wroclaw, 3).
droga(szczecin, warszawa, 4).

Reguła
droga(X, Y) :-
    droga(X, Y, _);
    droga(Y, X, _).

czas_przejazdu(X, Y, T) :-
    droga(X, Y, T).
czas_przejazdu(X, Y, T) :-
    droga(X, Z, T1),
    czas_przejazdu(Z, Y, T2),
    T is T1 + T2.

najkrotsza_trasa(X, Y) :-
    findall([Czas, Trasa], znajdz_trase(X, Y, [], Trasa, Czas), Trasy),
    Trasy \= [],
    minimalny_trasa(Trasy, [NajlepszyCzas, NajlepszaTrasa]),
    format('Najkrótsza trasa: ~w, czas: ~w~n', [NajlepszaTrasa, NajlepszyCzas]).

znajdz_trase(X, Y, Odwiedzone, Trasa, Czas) :-
    droga(X, Y, Czas),
    \+ member(Y, Odwiedzone),
    Trasa = [X, Y].
znajdz_trase(X, Y, Odwiedzone, Trasa, Czas) :-
    droga(X, Z, T1),
    \+ member(Z, Odwiedzone),
    znajdz_trase(Z, Y, [X|Odwiedzone], Podtrasa, T2),
    Trasa = [X|Podtrasa],
    Czas is T1 + T2.

minimalny_trasa([[Czas, Trasa]], [Czas, Trasa]).
minimalny_trasa([[Czas1, Trasa1]|Rest], [CzasMin, TrasaMin]) :-
    minimalny_trasa(Rest, [Czas2, Trasa2]),
    (Czas1 =< Czas2 ->
        (CzasMin = Czas1, TrasaMin = Trasa1);
        (CzasMin = Czas2, TrasaMin = Trasa2)).

Zapytanie
droga(warszawa, krakow).

Result
true

Zapytanie
czas_przejazdu(warszawa, poznan, T).

Result
T = 7

Zapytanie
najkrotsza_trasa(warszawa, poznan).

Result
Najkrótsza trasa: [warszawa,poznan], czas: 7
true
```

```python
#5
:- dynamic ksiazka/3.
:- dynamic ocena/3.
:- dynamic preferencje/2.

Fakty:
ksiazka('W pustyni i w puszczy', sienkiewicz, przygodowa).
ksiazka('Hobbit', tolkien, fantasy).
ksiazka('Pan Tadeusz', mickiewicz, poezja).
ksiazka('Zemsta', aleksander, komedia).
ksiazka('Wojna i pokój', tolstoy, historyczna).

ocena(uzytkownik1, 'W pustyni i w puszczy', 5).
ocena(uzytkownik1, 'Hobbit', 4).
ocena(uzytkownik1, 'Pan Tadeusz', 3).
ocena(uzytkownik2, 'Hobbit', 5).
ocena(uzytkownik2, 'Wojna i pokój', 4).

preferencje(uzytkownik1, fantasy).
preferencje(uzytkownik2, historyczna).

Reguły
polecane_ksiazki(Uzytkownik, Gatunek) :-
    preferencje(Uzytkownik, Gatunek),
    findall(Tytul, (ksiazka(Tytul, _, Gatunek), ocena(Uzytkownik, Tytul, Ocena), Ocena >= 4), Ksiazki),
    format('Polecane książki dla ~w w gatunku ~w: ~w', [Uzytkownik, Gatunek, Ksiazki]).

dodaj_ocene(Uzytkownik, Tytul, Ocena) :-
    assertz(ocena(Uzytkownik, Tytul, Ocena)).

usun_ocene(Uzytkownik, Tytul) :-
    retract(ocena(Uzytkownik, Tytul, _)).

Zapytanie
polecane_ksiazki(uzytkownik1, fantasy).

Result
Polecane książki dla uzytkownik1 w gatunku fantasy: [Hobbit]
```
